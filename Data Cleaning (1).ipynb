```python
# filter function = the  filter function returns an iterator where the items are filtered through a function to test , if the line is accepted return true or false.
ages = [5,16,18,19,20,32,21,30]
def myFunc(x):
    if(x>=18):
        return True
    else:
        return False
adult = list(filter(myFunc , ages))
print(adult)
```

```python
# Map function = the map() executes a specified function for each items in a iterable.
# The item is sent to the function as a perameter.

#example
```

```python
a = ['1' , '2' , '3' , '4' , '5']
b = list(map(int , a))
print(b)
b[2] = b[2] + 1
print(b[2])
```

```python
# exapmle
a = [1,2,3,4,5,6]
def squ(a):
    return a*a
square = list(map(squ , a))
print(square)
```

```python
# there is another simple method to do it

a = [1,3,4,5,6,76,7,8]
square = list(map(lambda x: x*x , a))
print(square)
```
